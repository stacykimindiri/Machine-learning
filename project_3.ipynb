```python
#checking for outliers(this are numerical columns)
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Calculate the number of outliers for each numerical column
outliers = df[numerical_columns].apply(lambda x: (x - x.mean()).abs() > 3 * x.std()).sum()

if outliers.empty:
    print("No outliers detected.")
else:
    print(f"Outliers:\n{outliers}")
```

This one was to check for outliers by calculation of z score and we can conclude that this data set has no outliers.
